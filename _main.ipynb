```python
import numpy as np
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
```

```python
img = plt.imread(r'data\yasuo.png')[:, :, :3]
plt.imshow(img)
plt.show()
```

```python
img = img / 255
X_img = np.reshape(img, (img.shape[0] * img.shape[1], 3))
```

```python
from models.kMeans import kMeans
from models.DBScan import DBScan
```

```python
idx, centroids = DBScan(X_img, k=16, num_iterations=10)
```

```python
X_recovered = centroids[idx, :]
X_recovered = np.reshape(X_recovered, img.shape)
```

```python
fig, ax = plt.subplots(1,2, figsize=(8,8))
plt.axis('off')

ax[0].imshow(img*255)
ax[0].set_title('Original')
ax[0].set_axis_off()

ax[1].imshow(X_recovered*255)
ax[1].set_title('Compressed with 16 colours')
ax[1].set_axis_off()
```

```python
# Extract the individual columns
x = X_img[:, 0]
y = X_img[:, 1]
z = X_img[:, 2]

# Set the figure size
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D scatter plot with colored points based on labels
scatter = ax.scatter(x, y, z, c=idx)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Data plot in 3D')

# Add a colorbar for the labels
cbar = fig.colorbar(scatter)
cbar.set_label('Labels')

# Show the plot
plt.show()
```
